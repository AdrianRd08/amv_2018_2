```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('Datos/encuesta_coders_limpio.xlsx')
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

```python
mm_scaler = MinMaxScaler()
std_scaler = StandardScaler()
pca = PCA(n_components=2)
mds = MDS(n_components=2)
```

```python
X = df.copy()
```

```python
mm_scaler.fit(X)
std_scaler.fit(X)
```

```python
Xmm = pd.DataFrame(mm_scaler.transform(X),columns=X.columns)
```

```python
Xs  = pd.DataFrame(std_scaler.transform(X),columns=X.columns)
```

```python
pca.fit(Xs)
print pca.explained_variance_ratio_.cumsum()
Xp = pd.DataFrame(pca.transform(Xs),columns=['p1','p2'])
```

```python
Xmds = pd.DataFrame(mds.fit_transform(Xmm.sample(frac=0.05)),columns=['d1','d2'])
```

```python
sns.lmplot(data=Xmds,x='d1',y='d2',fit_reg = False)
```

```python
sns.lmplot(data=Xp.sample(frac=0.05),x='p1',y='p2',fit_reg = False)
```
